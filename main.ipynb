```python
import requests
from bs4 import BeautifulSoup

def scrape(url):
    response = requests.get(url)
    response.encoding = 'utf-8'
    soup = BeautifulSoup(response.text, 'html.parser')

    title = soup.select_one('h1').text if soup.select_one('h1') else "ไม่มีหัวข้อ"
    text = soup.select_one('p').text if soup.select_one('p') else "ไม่มีเนื้อหา"
    link = soup.select_one('a').get('href') if soup.select_one('a') else "ไม่มีลิงก์"

    # ✅ ต้องคืนค่ากลับมา
    return title, text, link
```

```python
url = "https://thestandard.co/spending-grows-despite-card-closures/"   # เว็บภาษาไทยที่ใช้เป็นตัวอย่าง
title, text, link = scrape(url)
print("Title:", title)
print("Text:", text)
print("Link:", link)
```
